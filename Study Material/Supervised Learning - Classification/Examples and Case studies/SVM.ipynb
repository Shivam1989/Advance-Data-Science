```python
#Import xtrain from csv file
import pandas as pd
import numpy as np
xtrain = pd.read_csv("Khan_xtrain.csv", header = None)
del xtrain[0]
xtrain = xtrain[1:]
```

```python
#Import ytrain from csv file
ytrain = pd.read_csv("Khan_ytrain.csv", header = None)
del ytrain[0]
ytrain = ytrain[1:]
```

```python
xtrain[:10]
```

```python
from sklearn import svm
X = xtrain
y = ytrain
clf = svm.SVC(kernel='linear')
clf.fit(X, y)  
```

```python
#Import xtest from csv file
xtest = pd.read_csv("Khan_xtest.csv", header = None)
del xtest[0]
xtest = xtest[1:]
```

```python
#Import ytest from csv file
ytest = pd.read_csv("Khan_ytest.csv", header = None)
del ytest[0]
ytest = ytest[1:]
```

```python
#Make predictions on xtest dataset
pred = clf.predict(xtest)
pred
```

```python
#Performance on test dataset
pd.crosstab(pred,ytest[1],rownames=['pred'], colnames=['ytest'])
```
